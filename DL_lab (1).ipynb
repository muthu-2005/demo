```python
#colabcode
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow.keras import layers

# Load the dataset
(train_ds, val_ds, test_ds), metadata = tfds.load('tf_flowers', split=['train[:80%]', 'train[80%:90%]', 'train[90%:]'], with_info=True, as_supervised=True)

# Extract the number of classes and label names
num_classes = metadata.features['label'].num_classes
get_label_name = metadata.features['label'].int2str

# Display a sample image and its label
image, label = next(iter(train_ds))
_ = plt.imshow(image)
_ = plt.title(get_label_name(label))

# Image Preprocessing: Resizing and Rescaling
IMG_SIZE = 180
resize_and_rescale = tf.keras.Sequential([
    layers.Resizing(IMG_SIZE, IMG_SIZE),
    layers.Rescaling(1./255)
])

# Display the resized image
result = resize_and_rescale(image)
_ = plt.imshow(result)
print("Min and max pixel values:", result.numpy().min(), result.numpy().max())

# Data Augmentation
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(0.2),
])

# Create batches and apply data augmentation only to the training set
def prepare(ds, shuffle=False, augment=False):
    ds = ds.map(lambda x, y: (resize_and_rescale(x), y), num_parallel_calls=tf.data.AUTOTUNE)
    if shuffle:
        ds = ds.shuffle(1000)
    ds = ds.batch(32)
    if augment:
        ds = ds.map(lambda x, y: (data_augmentation(x, training=True), y), num_parallel_calls=tf.data.AUTOTUNE)
    return ds.prefetch(buffer_size=tf.data.AUTOTUNE)

train_ds = prepare(train_ds, shuffle=True, augment=True)
val_ds = prepare(val_ds)
test_ds = prepare(test_ds)

# Build the model
model = tf.keras.Sequential([
    layers.InputLayer(input_shape=(IMG_SIZE, IMG_SIZE, 3)),  # Fixed input size
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(num_classes)  # Output layer with num_classes neurons
])

# Compile the model
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# Train the model
epochs = 5
history = model.fit(train_ds, validation_data=val_ds, epochs=epochs)

# Evaluate the model on the test dataset
loss, acc = model.evaluate(test_ds)
print("Accuracy", acc)

# Optional: Visualize some augmentations
def visualize(original, augmented):
    fig = plt.figure()
    plt.subplot(1, 2, 1)
    plt.title('Original image')
    plt.imshow(original)
    plt.subplot(1, 2, 2)
    plt.title('Augmented image')
    plt.imshow(augmented)

flipped = tf.image.flip_left_right(image)
visualize(image, flipped)
grayscaled = tf.image.rgb_to_grayscale(image)
visualize(image, tf.squeeze(grayscaled))
saturated = tf.image.adjust_saturation(image, 3)
visualize(image, saturated)
bright = tf.image.adjust_brightness(image, 0.4)
visualize(image, bright)
cropped = tf.image.central_crop(image, central_fraction=0.5)
visualize(image, cropped)
rotated = tf.image.rot90(image)
visualize(image, rotated)
```

```python
#juptercode
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np

# Define training input (X) and output (Y) data
# Each input maps to an output
X = tf.constant([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=tf.float32)
Y = tf.constant([[0], [1], [1], [0]], dtype=tf.float32)

# Create a Sequential model
model = keras.Sequential()

# Add layers
model.add(layers.Dense(
    4,  # Number of neurons
    input_dim=2,  # Input dimension for the first layer
    activation='relu'  # ReLU activation function
))
model.add(layers.Dense(
    1,  # Number of neurons for output
    activation='sigmoid'  # Sigmoid activation function for binary classification
))

# Compile the model
model.compile(
    loss='binary_crossentropy',  # Use binary cross-entropy for binary classification
    optimizer='adam',  # Adam optimizer
    metrics=['binary_accuracy']  # Track binary accuracy
)

# Function to predict the output for two inputs
def cleanPredict(a, b):
    # Prepare input as a 2D tensor
    inputTens = tf.constant([[a, b]], dtype=tf.float32)
    # Make prediction
    prediction = model.predict(inputTens, verbose=0)[0][0]
    return round(prediction) == 1

# Display random prediction before training
print("Prediction before training:")
print(f"cleanPredict(1, 0): {cleanPredict(1, 0)}")

# Train the model
print("\nTraining the model...")
history = model.fit(
    X,  # Input data
    Y,  # Output data
    epochs=20,  # Number of training iterations
    verbose=1  # Verbosity level
)

# Display predictions after training
print("\nPredictions after training:")
print(f"cleanPredict(1, 0): {cleanPredict(1, 0)}")  # Should yield True
print(f"cleanPredict(0, 0): {cleanPredict(0, 0)}")  # Should yield False
print(f"cleanPredict(1, 1): {cleanPredict(1, 1)}")  # Should yield False
print(f"cleanPredict(0, 1): {cleanPredict(0, 1)}")  # Should yield True
```

**Exp 3(Classify cat and dog using CNN)**

```python
#both colab and jupytercode
import tensorflow as tf
from tensorflow.keras.preprocessing import image_dataset_from_directory
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import os

# Download Cats vs Dogs dataset
_URL = 'https://storage.googleapis.com/mledu-datasets/cats_and_dogs_filtered.zip'
path_to_zip = tf.keras.utils.get_file('cats_and_dogs.zip', origin=_URL, extract=True)

# Define directories for train and validation datasets
base_dir = os.path.join(os.path.dirname(path_to_zip), 'cats_and_dogs_filtered')
train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')

# Load datasets
train_ds = image_dataset_from_directory(
    train_dir,
    image_size=(180, 180),
    batch_size=32,
    label_mode='int',  # Use 'int' for sparse labels (0 or 1 for Cats and Dogs)
)

val_ds = image_dataset_from_directory(
    validation_dir,
    image_size=(180, 180),
    batch_size=32,
    label_mode='int',
)

# Normalize pixel values to [0, 1]
normalization_layer = layers.Rescaling(1./255)

# Create the CNN model
model = models.Sequential([
    normalization_layer,
    layers.Conv2D(16, 3, padding='same', activation='relu', input_shape=(180, 180, 3)),
    layers.MaxPooling2D(),
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(1, activation='sigmoid')  # Sigmoid for binary classification (Cat vs Dog)
])

# Compile the model
model.compile(
    optimizer=Adam(),
    loss='binary_crossentropy',  # Binary crossentropy for binary classification
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10
)

# Evaluate the model on the validation set
val_loss, val_acc = model.evaluate(val_ds)
print(f'Validation accuracy: {val_acc*100:.2f}%')

# Plot training & validation accuracy and loss
plt.figure(figsize=(12, 4))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy')
plt.legend()

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss')
plt.legend()

plt.show()
```

**Exp 7(Object Detection CNN)**

```python
#colabcode and jupyter

import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
from keras.models import Model
from keras.layers import Conv2D, Activation, Dense, Flatten, Input
from tensorflow.keras.optimizers import SGD

# Load CIFAR-10 dataset
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Normalize inputs to range [0, 1]
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# One-hot encode the labels
Y_train = to_categorical(y_train, 10)
Y_test = to_categorical(y_test, 10)

# Create dummy bounding boxes for training (random values)
def generate_random_boxes(batch_size=32):
    # CIFAR-10 images are 32x32, so bounding box coordinates are within this range
    x_min = np.random.randint(0, 16, batch_size)  # x_min from 0 to 16
    y_min = np.random.randint(0, 16, batch_size)  # y_min from 0 to 16
    x_max = np.random.randint(16, 32, batch_size)  # x_max from 16 to 32
    y_max = np.random.randint(16, 32, batch_size)  # y_max from 16 to 32
    return np.stack([x_min, y_min, x_max, y_max], axis=-1)

# Simulate bounding box data for the training and testing sets
bbox_train = generate_random_boxes(X_train.shape[0])
bbox_test = generate_random_boxes(X_test.shape[0])

# Define a simpler object detection model
def simple_object_detection_model():
    input_layer = Input(shape=(32, 32, 3))  # Input layer

    x = Conv2D(32, (3, 3), padding='same')(input_layer)
    x = Activation('relu')(x)

    x = Conv2D(64, (3, 3), padding='same')(x)
    x = Activation('relu')(x)

    x = Flatten()(x)
    fc1 = Dense(256, activation='relu')(x)  # Fully connected layer for both bbox and class prediction

    # Two output layers
    bbox_output = Dense(4, activation='linear', name='bbox_output')(fc1)  # 4 values for bounding box (x_min, y_min, x_max, y_max)
    class_output = Dense(10, activation='softmax', name='class_output')(fc1)  # 10 values for class prediction

    # Create the model with two outputs
    model = Model(inputs=input_layer, outputs=[bbox_output, class_output])
    return model

# Define optimizer and compile model
learning_rate = 0.01
weight_decay = 1e-6
momentum = 0.9
epochs = 5
batch_size = 800  # Batch size

sgd = SGD(learning_rate=learning_rate, decay=weight_decay, momentum=momentum, nesterov=True)  # Optimizer
model = simple_object_detection_model()

# Compile model with loss functions for both outputs
model.compile(loss={'bbox_output': 'mean_squared_error', 'class_output': 'categorical_crossentropy'},
              optimizer=sgd, metrics={'class_output': 'accuracy'})

model.summary()

# Train the model with both class labels and bounding box labels
model.fit(X_train, {'bbox_output': bbox_train, 'class_output': Y_train},
          validation_data=(X_test, {'bbox_output': bbox_test, 'class_output': Y_test}),
          epochs=epochs, batch_size=batch_size, verbose=1)

# Make predictions
batch = X_test[100:109]
bbox_pred, class_pred = model.predict(batch)

# Print bounding boxes and class predictions for 9 test images
for i in range(9):
    print(f"Image {i + 1}: Predicted Bounding Box: {bbox_pred[i]}, Predicted Class: {np.argmax(class_pred[i])}")

    # Plot the image and bounding box
    plt.subplot(3, 3, i + 1)
    img = batch[i]
    plt.imshow(img)
    x_min, y_min, x_max, y_max = bbox_pred[i]
    plt.gca().add_patch(plt.Rectangle((x_min, y_min), x_max - x_min, y_max - y_min,
                                      linewidth=2, edgecolor='r', facecolor='none'))
    plt.title(f"Class: {np.argmax(class_pred[i])}")

plt.show()
```

**Exp 10(GAN)**

```python
#both colab and jupytercode

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Reshape, Flatten, LeakyReLU, BatchNormalization, Conv2D

# Load and preprocess MNIST dataset
(X_train, _), (_, _) = tf.keras.datasets.mnist.load_data()
X_train = (X_train - 127.5) / 127.5  # Normalize to [-1, 1]
X_train = np.expand_dims(X_train, axis=-1)

img_shape, latent_dim = (28, 28, 1), 100

# Discriminator
def build_discriminator():
    model = Sequential([
        Conv2D(64, 3, 2, padding='same', input_shape=img_shape),
        LeakyReLU(0.2),
        Conv2D(128, 3, 2, padding='same'),
        LeakyReLU(0.2),
        Flatten(),
        Dense(1, activation='sigmoid')
    ])
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

# Generator
def build_generator():
    model = Sequential([
        Dense(256, input_dim=latent_dim),  # Define input_dim here to avoid None shape
        LeakyReLU(0.2),
        BatchNormalization(momentum=0.8),
        Dense(512),
        LeakyReLU(0.2),
        BatchNormalization(momentum=0.8),
        Dense(1024),
        LeakyReLU(0.2),
        BatchNormalization(momentum=0.8),
        Dense(np.prod(img_shape), activation='tanh'),
        Reshape(img_shape)
    ])
    return model

# GAN
def build_gan(generator, discriminator):
    discriminator.trainable = False
    model = Sequential([generator, discriminator])
    model.compile(loss='binary_crossentropy', optimizer='adam')
    return model

# Training function
def train_gan(epochs, batch_size=128, sample_interval=1000):
    real, fake = np.ones((batch_size, 1)), np.zeros((batch_size, 1))
    for epoch in range(epochs):
        # Train discriminator
        idx = np.random.randint(0, X_train.shape[0], batch_size)
        real_imgs = X_train[idx]
        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        gen_imgs = generator.predict(noise)
        d_loss_real = discriminator.train_on_batch(real_imgs, real)
        d_loss_fake = discriminator.train_on_batch(gen_imgs, fake)
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

        # Train generator
        g_loss = gan.train_on_batch(noise, real)

        if epoch % sample_interval == 0:
            sample_images(epoch)
        print(f"{epoch} [D loss: {d_loss[0]}, acc.: {100*d_loss[1]}] [G loss: {g_loss}]")

# Generate and visualize sample images
def sample_images(epoch, grid_size=5):
    noise = np.random.normal(0, 1, (grid_size * grid_size, latent_dim))
    gen_imgs = generator.predict(noise)
    gen_imgs = 0.5 * gen_imgs + 0.5  # Rescale to [0, 1]

    fig, axs = plt.subplots(grid_size, grid_size)
    for i in range(grid_size):
        for j in range(grid_size):
            axs[i, j].imshow(gen_imgs[i * grid_size + j, :, :, 0], cmap='gray')
            axs[i, j].axis('off')
    plt.show()

# Models and training
discriminator = build_discriminator()
generator = build_generator()
gan = build_gan(generator, discriminator)

train_gan(epochs=100, batch_size=64, sample_interval=1000)

```

**Exp 11(LSTM)**

```python
#both colab and jupytercode
import tensorflow as tf
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Embedding, Dense, Add
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# Step 1: Load Pre-trained CNN Model (InceptionV3)
base_model = InceptionV3(weights='imagenet')
cnn_model = Model(inputs=base_model.input, outputs=base_model.layers[-2].output)  # Extract features

def preprocess_image(image_path):
    img = tf.keras.utils.load_img(image_path, target_size=(299, 299))
    img = tf.keras.utils.img_to_array(img)
    img = tf.keras.applications.inception_v3.preprocess_input(img)
    return np.expand_dims(img, axis=0)

def extract_features(image_path):
    """Extracts features from an image using InceptionV3."""
    img = preprocess_image(image_path)
    features = cnn_model.predict(img)
    return features.reshape(2048)

# Step 2: Prepare Tokenizer and Dummy Captions
captions = ["a dog in a park", "a cat on a sofa", "a person on a bicycle"]
tokenizer = Tokenizer()
tokenizer.fit_on_texts(captions)
vocab_size = len(tokenizer.word_index) + 1

# Tokenize captions
sequences = tokenizer.texts_to_sequences(captions)
max_caption_len = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_caption_len, padding='post')

# Step 3: Build Image Captioning Model
def create_model(vocab_size, max_caption_len):
    # Image features input
    image_input = Input(shape=(2048,))
    image_features = Dense(256, activation='relu')(image_input)

    # Caption input
    caption_input = Input(shape=(max_caption_len,))
    caption_embedding = Embedding(vocab_size, 256, mask_zero=True)(caption_input)
    caption_lstm = LSTM(256, return_sequences=True)(caption_embedding)  # Return sequences for each timestep

    # Combine both inputs
    combined = Add()([image_features, caption_lstm])
    output = Dense(vocab_size, activation='softmax')(combined)  # Output for each word prediction

    model = Model(inputs=[image_input, caption_input], outputs=output)
    model.compile(optimizer='adam', loss='categorical_crossentropy')
    return model

model = create_model(vocab_size, max_caption_len)

# Step 4: Prepare Image Features and Target Sequences
# Extract features for the images
image_features = np.array([extract_features('/content/cat.jpg')] * len(captions))  # Using the same image for all captions

# Prepare the target sequences (shifted one step for each word in the sequence)
target_sequences = np.zeros_like(padded_sequences)
target_sequences[:, :-1] = padded_sequences[:, 1:]  # Shift the sequence by one

# Convert target sequences to one-hot encoding
target_sequences_one_hot = tf.keras.utils.to_categorical(target_sequences, num_classes=vocab_size)

# Step 5: Train the Model
model.fit([image_features, padded_sequences], target_sequences_one_hot, epochs=5)
```

```python
#both colab and jupytercode
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np

# Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalize the data to values between 0 and 1
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Reshape the data to include a channel dimension (needed for CNNs)
X_train = np.expand_dims(X_train, axis=-1)
X_test = np.expand_dims(X_test, axis=-1)

# Convert labels to categorical (one-hot encoding)
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)

# Visualize a few training examples
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(X_train[i].reshape(28, 28), cmap='gray')
    plt.axis('off')
plt.show()


# Build the CNN model
model = models.Sequential()

# First convolutional layer
model.add(layers.Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))

# Second convolutional layer
model.add(layers.Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))

# Flatten the output from the convolutional layers
model.add(layers.Flatten())

# Fully connected layer
model.add(layers.Dense(128, activation='relu'))

# Output layer with 10 units (one for each class), using softmax activation for classification
model.add(layers.Dense(10, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, validation_split=0.2, epochs=10, batch_size=128, verbose=2)
```

```python
#both colab and jupytercode
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# Step 1: Load SMILES Data
# Example SMILES dataset
smiles_data = [
    "CCO",  # Ethanol
    "C(C(=O)O)N",  # Glycine
    "CC(C)O",  # Isopropanol
    "CCC",  # Propane
    "C1=CC=CC=C1"  # Benzene
]

# Step 2: Tokenize SMILES Data
tokenizer = Tokenizer(char_level=True)  # Character-level tokenization for SMILES
tokenizer.fit_on_texts(smiles_data)
vocab_size = len(tokenizer.word_index) + 1

# Convert SMILES to sequences
sequences = tokenizer.texts_to_sequences(smiles_data)
max_seq_length = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_seq_length, padding='post')

# Prepare input and output data for sequence prediction
X = padded_sequences[:, :-1]  # Input sequence
y = padded_sequences[:, 1:]  # Output sequence (next character)
y = tf.keras.utils.to_categorical(y, num_classes=vocab_size)

# Step 3: Build a Simple LSTM Model
model = Sequential([
    Embedding(vocab_size, 128, input_length=max_seq_length - 1),
    LSTM(256, return_sequences=True),
    Dense(vocab_size, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Corrected line: use X (not x) for input data
history = model.fit(X, y, validation_split=0.2, epochs=10, batch_size=128, verbose=2)
```
