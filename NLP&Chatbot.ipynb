Ex. No.1 TOKENIZE AND TAG SOME TEXT


```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import TweetTokenizer
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
import spacy
nltk.download('punkt')
sample_text = "this is a text ready to tokenize"
tokens = word_tokenize(sample_text)

def print_text(original, tokenized):
    print("Original Text: ", original)
    print("Tokenized Text: ", tokenized)

print_text(sample_text, tokens)

tweet_tokenizer = TweetTokenizer()
sample_text = "This is a tweet @jack #NLP"
tokens = tweet_tokenizer.tokenize(sample_text)
print_text(sample_text, tokens)

sample_text = "This is a sentence. This is another one!\nAnd this is thelast one."
sentences = sent_tokenize(sample_text)
print_text(sample_text, sentences)

nltk.download("stopwords")
stopwords_ = set(stopwords.words("english"))

sample_text = "this is a sample text"
tokens = sample_text.split()

clean_tokens = [t for t in tokens if not t in stopwords_]
clean_text = " ".join(clean_tokens)
print_text(sample_text, clean_text)

nlp = spacy.load("en_core_web_sm")
sample_text = "this is a text ready to tokenize"
doc = nlp(sample_text)
tokens = [token.text for token in doc]
print_text(sample_text, tokens)

nlp = spacy.load("en_core_web_sm")
sample_text = "This is a sentence. This is another one!\nAnd this is the last one."
doc = nlp(sample_text)
sentences = [sentence.text for sentence in doc.sents]
print_text(sample_text, sentences)
```

Ex.No.2 RESUME SCREENING USING NLP

```python
import spacy
def resume_screening(resume_text):

  nlp = spacy.load("en_core_web_sm")
  doc = nlp(resume_text)
  desired_skills = ["python", "machine learning", "data analysis", "communication"]
  extracted_skills = [token.text.lower() for token in doc if token.text.lower() in desired_skills]
  years_of_experience = extract_years_of_experience(doc)
  meets_criteria = (
  "python" in extracted_skills
  and ("machine learning" in extracted_skills or "communication" in extracted_skills)
  and years_of_experience >= 2
  )

  print("Extracted Skills:", extracted_skills)
  print("Years of Experience:", years_of_experience)
  print("Meets Screening Criteria:", meets_criteria)

def extract_years_of_experience(doc):



  for ent in doc.ents:
    if ent.label_ == "DATE" or (ent.text.isdigit() and "year" in ent.text.lower()):
      for token in ent:
        if token.pos_ == "NUM" and token.dep_ == "nummod":
            return int(token.text)
  return 0

resume_text = """
I am a Python developer with 3 years of experience. My expertise includes machine learning, data
analysis, and effective communication.
"""

resume_screening(resume_text)
```

Ex.No.3 DEVELOPMENT OF SENTIMENT ANALYSIS WITH PYTHON

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
analyzer = SentimentIntensityAnalyzer()

text = "I am going to die."

sentiment_scores = analyzer.polarity_scores(text)

compound_score = sentiment_scores['compound']
if compound_score >= 0.05:
  sentiment = "Positive"
elif compound_score <= -0.05:
  sentiment = "Negative"
else:
  sentiment = "Neutral"

print(f"Sentiment: {sentiment}")
print(f"Compound Score: {compound_score}")
```

Ex.No: 4 DEVELOP KEYWORD EXTRACTION WITH PYTHON

```python
from rake_nltk import Rake
from wordcloud import WordCloud
import matplotlib.pyplot as plt

rake = Rake()

input_text = '''
NLP stands for Natural Language Processing.
It is the branch of Artificial Intelligence that gives the ability to machine understand
and process human languages. Human languages can be in the form of text or audio format.
Natural Language Processing started in 1950 When Alan Mathison Turing published
an article in the name Computing Machinery and Intelligence.
It is based on Artificial intelligence. It talks about automatic interpretation and
generation of natural language.
As the technology evolved, different approaches have come to deal with NLP tasks.
'''

rake.extract_keywords_from_text(input_text)
keywords = rake.get_ranked_phrases()

print(keywords)
wordcloud = WordCloud().generate(' '.join(keywords))

plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Ex.No:7 CHATBOT IMPLEMENTATION USING NLP

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
def chatbot(user_input, model, tokenizer, max_length=100):
  input_ids = tokenizer.encode(user_input, return_tensors="pt")
  output = model.generate(input_ids, max_length=max_length, num_beams=5,
  no_repeat_ngram_size=2, top_k=50, top_p=0.95)
  response = tokenizer.decode(output[0], skip_special_tokens=True)
  return response
def chat_with_user(model, tokenizer):
  print("Hi, I'm your chatbot. Type 'quit' to exit.")
  while True:
    user_input = input("You: ")
    if user_input.lower() == 'quit':
      print("Goodbye!")
      break
    response = chatbot(user_input, model, tokenizer)
    print("Bot:", response)

chat_with_user(model, tokenizer)
```

Ex.No:8 IMPLEMENT OF NEXT WORD PREDICTION MODEL

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
import numpy as np
seed_text = "Pizza have different"
next_words = 5
for _ in range(next_words):
  token_list = tokenizer.encode(seed_text, return_tensors="pt")
  predicted_probs = model(token_list).logits[:, -1, :]
  predicted_word_index = torch.argmax(predicted_probs).item()
  predicted_word = tokenizer.decode([predicted_word_index])
  seed_text += " " + predicted_word
print("Next predicted words:", seed_text)
```

Exp.No:9 COLLEGE ENQUIRY CHATBOT


```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
class CollegeEnquiryChatbot:
  def __init__ (self):
    self.model = GPT2LMHeadModel.from_pretrained("gpt2")
    self.tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
  def generate_response(self, user_input, max_length=100):
    input_ids = self.tokenizer.encode(user_input, return_tensors="pt")
    output = self.model.generate(input_ids, max_length=max_length, num_beams=5,no_repeat_ngram_size=2, top_k=50, top_p=0.95)
    response = self.tokenizer.decode(output[0], skip_special_tokens=True)
    return response
def chat_with_college_enquiry_bot():
  college_bot = CollegeEnquiryChatbot()
  print("College Enquiry Chatbot: Hi! I'm here to help you with information about the college. Type 'quit' to exit.")
  while True:
    user_input = input("You: ")
    if user_input.lower() == 'quit':
      print("College Enquiry Chatbot: Goodbye!")
      break
  response = college_bot.generate_response(user_input)
  print("College Enquiry Chatbot:", response)

chat_with_college_enquiry_bot()
```
